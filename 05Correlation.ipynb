# Correlation


**상관계수 (Correlation coefficient)**

두 변수 x, y의 상관관계의 강도를 -1 ~ 1 사이의 수치로 표현.

상관계수의 값이


*   1에 가까울 수록 : 양의 상관관계 -> 변수 x가 증가하면 변수 y도 증가
*   -1에 가까울 수록 : 음의 상관관계 -> 변수 x가 커질수록 변수 y는 감소
*   0에 가까울 수록 : 상관관계가 없음 -> 변수 x가 변수 y에 영향을 미치지 않음

상관분석 (Correlation analysis, 상관분석)

두 변수 사이 관계의 강도와 방향을 파악하는 통계 기법. 상관계수만 가지고 두 변수 사이의 상관성이 있는지 없는지 판단할 수는 없음. 두 번수에 선형 상관관계가 아닌 다른 상관관계가 있을 수 있음.

1. Pearson correlation analysis : 피어슨 상관분석은 상관계수로 '만' 두 변수 사이의 선형 관계(linear relationship)을 파악하는 일반적인 상관분석 방법
2. Spearman correlation analysis : 스피어만 상관분석은 변수의 '순위' 만을 따져 두 변수의 단조 관계성(monotonic relationship)을 사용하여 얼마나 잘 설명될 수 있는지를 평가.
3. Kendall correlation analysis : 켄달 상관분석은 스피어만 상관분석과 비슷. but, 전체 데이터가 부족할 때 많이 사용



```python
import pandas as pd

# 영어 리스닝, 총 점수
engListening = [30, 60, 90]
engScore = [70, 80, 90]

# 리스트 데이터 넣을 columns 이름으로 딕셔너리 생성
data = {'engListening':engListening, 'engScore':engScore}
# 딕셔너리로 DataFrame 생성
df = pd.DataFrame(data)

# coef = 상관계수 pearson 메소드로 생성
coef = df.corr(method='pearson')
print(coef)

import matplotlib.pyplot as plt

# scatter로 그래프 그리기
plt.scatter(df['engListening'], df['engScore'])
plt.show()
```

```python
engListening = [30, 60, 90, 31, 32, 69, 92, 99]
engScore = [70, 80, 90, 70, 71, 85, 90, 92]

data2 = {'engListening':engListening, 'engScore':engScore}
df2 = pd.DataFrame(data2)

plt.scatter(df2['engListening'], df2['engScore'])
plt.show()

print('<< Pearson >>')
coef = df2.corr(method='pearson')
print(coef)
print('\n<< Spearnman >>')
spearmanCoef = df2.corr(method='spearman')
print(spearmanCoef)  # 스피어만 = 커지냐 작아지냐만 따짐
print('\n<< Kendall >>')
kendallCoef = df2.corr(method='kendall')
print(kendallCoef)
```

```python
engListening = [30, 60, 90, 31, 32, 69, 92, 99]
engReading = [40, 45, 60, 20, 15, 70, 60, 80]
engClass = [60, 120, 120, 60, 60, 180, 120, 120]
engScore = [70, 80, 90, 70, 71, 85, 90, 92]

data3 = {'engListening':engListening, 'engReading':engReading, 'engClass':engClass,
         'engScore':engScore}
df3 = pd.DataFrame(data3)

plt.scatter(df3['engListening'], df3['engScore'])
plt.scatter(df3['engReading'], df3['engScore'])
plt.scatter(df3['engClass'], df3['engScore'])
plt.show()

print('<< Pearson >>')
coef = df3.corr(method='pearson')
print(coef)
print('\n<< Spearnman >>')
spearmanCoef = df3.corr(method='spearman')
print(spearmanCoef)  # 스피어만 = 커지냐 작아지냐만 따짐
print('\n<< Kendall >>')
kendallCoef = df3.corr(method='kendall')
print(kendallCoef)
```

```python
data = {'영어': [4, 2, 1, 3, 10, 8, 9, 7, 6, 5],
        '수학': [2, 1, 3, 4, 8, 7, 10, 5, 9, 6]}
df = pd.DataFrame(data)

dataG = {'영어': [84, 72, 71, 83, 98, 94, 96, 92, 86, 85],
         '수학': [82, 81, 83, 84, 97, 96, 99, 87, 98, 90]}
dfG = pd.DataFrame(dataG)

plt.scatter(df['영어'], df['수학'])
plt.show()
plt.scatter(dfG['영어'], dfG['수학'])
plt.show()

print('<< Pearson >>')
coef = df.corr(method='pearson')
print(coef)
coefG = dfG.corr(method='pearson')
print(coefG)

print('\n<< Spearnman >>')
spearmanCoef = df.corr(method='spearman')
print(spearmanCoef)  # 스피어만 = 커지냐 작아지냐만 따짐
print('\n<< Spearnman >>')
spearmanCoefG = dfG.corr(method='spearman')
print(spearmanCoefG)
```
