평면의 두 점의 좌표가 주어져 있을 때 두 점을 지나는 직선의 방정식을 출력하는 프로그램을 작성

```python
def funcX(p1, p2):
  a = (p1[1] - p2[1] / p1[0] - p2[0])
  b = p1[1] - a*p1[0]
  return a, b

a, b = funcX((1, 2), (2, 4))
print(f"y = {a}x + {b}")
```

```python
mask = np.array([0, 1, 1, 0], dtype=bool) # array Mask 생성
print('mask\n', mask)

data = np.random.randn(4, 2)
print('\ndata 원본\n', data)

# mask는 0, 1, 1, 0 이므로
# data[mask]는 두 번째, 세 번째 데이터만 True로 남고
print('\n[0, 1, 1, 0] masking\n', data[mask])
# data[~mask]는 첫 번째, 네 번째 데이터만 False의 역 = True 로 남는다.
print('\ninverse masking\n', data[~mask])

# 0보다 큰 데이터 추출
posit = data[data > 0]
print(' > 0\n', posit)

over1 = data[1][data[1] > 0]
print('data[1] and > 0\n', over1)
```

```python
origin = np.arange(6)
print('origin\n', origin)

copy = origin
print('copy\n', copy)

deepCopy = origin.copy()
print('deepCopy\n', deepCopy)

# 원본이 변형되면 일반 copy와 원본은 값이 변경
# deepCopy의 데이터는 변화 X
origin[1] = 10
print('\nchanging origin data\n')
print('origin\n', origin)
print('copy\n', copy)
print('deepCopy\n', deepCopy)

# 원본을 복사한 copy 값이 변경되면?
# 원본 array의 데이터 값도 변경된다.
copy[2] = 20
print('\nchanging copy data\n')
print('origin\n', origin)
print('copy\n', copy)
print('deepCopy\n', deepCopy)

# deepCopy는 데이터가 독립적
# 원본이 바뀌어도 deepCopy는 변화 X
# deepCopy가 바뀌어도 원본에 변화 X
deepCopy[3] = 30
print('\nchanging deepCopy data\n')
print('origin\n', origin)
print('copy\n', copy)
print('deepCopy\n', deepCopy)
```

# Data sort / argument sort

```python
a = np.array([3, 2, 5, 1, 4])
print('origin data\n', a)
print('= 0 1 2 3 4 =')

print('\nsorting data\n', np.sort(a)) # np.sort(a)로 값을 정렬
print('after np.sort\n', a) # 원본 데이터에는 변화 X
print('arguments sort\n', np.argsort(a)) # np.argsort(a) => 정렬된 argument 번호로 값을 출력해줌
print('after np.argsort\n', a) # 여전히 원본 데이터는 변화 X

print('argsort method\n', a.argsort())
print('after argsort()\n', a)
a.sort()
print('after sort()\n', a) # a에 직접 a.sort()는 a 원본 데이터 값도 정렬된 상태로 변형시킴
```
