```python
# 점을 통해 a(기울기), b(절편) 값 획득
def findFuncX(p1, p2):
  a = (p1[1] - p2[1] / p1[0] - p2[0])
  b = p1[1] - a*p1[0]
  return a, b

# a, b, x 값을 통해 y 값 획득
def funcX(a, b, x):
  y = a*x + b
  return y

a, b = findFuncX((1, 2), (2, 4))
print(a, b)

y = funcX(a, b, 7)
print(y)

y = [funcX(a, b, x) for x in range(100)]
print(y)

import matplotlib
import matplotlib.pyplot as plt

plt.plot(y)
plt.show()
```

# Multiple Figure & Subplot

 plt.figure(#, figsize=(#, #), dpi=#, facecolor='', edgecolor='', frameon=True/False)

 where

 #: figure number

 figsize: gure size in inch. using (width, hight) tuple

 dpi: dot per inch. default ▶ 100

 facecolor: using color name in string or RGB tuple.

 edgecolor: using color name in string or RGB tuple.

 frameon: True/False

 ex) plt.figure(1, figsize=(5, 8), dpi=300, facecolor='blue',
 edgecolor='black', frameon=True)

```python
def f(t): # 익스퍼넨셜 * 코사인 값을 t 값을 통해 생성
  return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1) # 0~5까지 0.1씩
t2 = np.arange(0.0, 5.0, 0.02) # 0~5까지 0.02씩

plt.figure(1) # plot 넣을 figure 생성
plt.subplot(211) # 211 위치에 그래프
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')
# t1 값을 통해 파란 원
# t2 값을 통해 검은색 선으로

plt.subplot(212) # 212 위치에 그래프
plt.plot(t2, np.cos(2*np.pi*t2), 'r--') # t2값으로 cos(2파이*t2) 빨간 점선
plt.show()
```

```python
x = np.arange(0., 5., .01)
y = np.cos(2*np.pi*x)

plt.figure(1, figsize=(8, 8))
plt.subplot(221)
plt.plot(x, y)
plt.title('fig 1')

plt.subplot(222)
plt.plot(x, -y)
plt.grid(True) # 십자선 grid 추가

plt.subplot(223)
plt.plot(x, y + 1)
plt.title('fig 3')

plt.subplot(224)
plt.plot(x, y**2)
plt.title('fig 4')

plt.show()
```

# Histogram

```python
np.random.seed(19680801) # 랜덤값 seed 설정
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

plt.hist(x, bins=100) # bins = 구간을 얼마나 나눌 것인가
# n, bins, patches = plt.hist(x, bins=100, histtype='stepfilled', facecolor='g', alpha=0.75) alpha = 투명도

plt.xlabel('Smarts', fontsize=18, color='red')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(60, 250, r'$\mu=100,\ \sigma=15$') # (60, 250) 위치에 mu, sigma text 삽입
plt.axis([40, 160, 0, 350]) # x, y축 시작과 끝 설정
plt.grid(True) # 십자선 추가
plt.show()
```

```python
ax = plt.subplot(111)

t = np.arange(0.0, 5.0, 0.01)
s = t*np.cos(2*np.pi*t)

plt.plot(t, s, lw=2)

plt.annotate('turning point', xy=(2,2), xytext=(3,5), arrowprops=dict(facecolor='black', shrink=0.05), )
# (2,2) 위치에 표시 / 해당하는 설명 text는 (3,5)에 표시 / arrowprops => 화살표로 표시, black 색상

plt.ylim(-8, 8) # y값 리미트 설정해서 범위를 -8~8까지
plt.show()
```

```python
fig = plt.figure(1, figsize=(5,5))
ax = plt.subplot(111)
bbox_props = dict(boxstyle="round", fc="w", ec="0.5", alpha=0.9)
ax.text(-2, -2, "Sample", ha="center", va="center", size=20, bbox=bbox_props)
# ax.text(0, 0, "Another", va="center", rotation=45, size=15, bbox=bbox_props)

ax.set_xlim(-4, 2) #x, y범위 설정 xlimit / ylimit
ax.set_ylim(-4, 2)

plt.show()
```

```python
fig = plt.figure(1, figsize=(5,5)) # figure 1 생성, 사이즈 설정
ax = plt.subplot(111) # ax가 그려질 plot 1개

bbox_props = dict(boxstyle="rarrow", fc="w", ec="0.5") # bbox에 들어갈 prop 미리 설정
t = ax.text(0, 0, "Cool", rotation=45, size=15, bbox=bbox_props) # Cool 텍스트, 기울기, 크기, bbox prop 설정 입력

ax.set_xlim(-4, 4) # x, y 리미트 설정
ax.set_ylim(-4, 4)

plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(1, figsize=(3,3))

ax = plt.subplot(111)
# Test 텍스트, xy 0.2값 지점 좌표 지정, text 좌표 지정, 화살표 설정 (-|>, arc3 사용)
ax.annotate("Test", xy=(0.2, 0.2), xycoords='data', xytext=(0.8, 0.8),
           textcoords='data', arrowprops=dict(arrowstyle="-|>", connectionstyle="arc3"))


plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(1, figsize=(3,3))
ax = plt.subplot(111)

# x,y값 (화살표가 가리킬 지점) 0.2 0.2
# 텍스트 0.8 0.8 지점, 사이즈
# 화살표 설정
ax.annotate("Test", xy=(0.2, 0.2), xycoords='data',
           xytext=(0.8, 0.8), textcoords='data', size=20, va="center", ha="center",
           arrowprops=dict(arrowstyle="simple", connectionstyle="arc3, rad=-0.2"))
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(1, figsize=(3,3))
# bbox 설정 (둥근상자 round4, fc = 상자 색)
bbox_props = dict(boxstyle="round4", fc="w")

ax = plt.subplot(111)
ax.annotate("Test", xy=(0.2, 0.2), xycoords='data',
           xytext=(0.8, 0.8), textcoords='data', size=20, va="center", ha="center", bbox=bbox_props,
           arrowprops=dict(arrowstyle="-|>",
           connectionstyle="arc3, rad=-0.2", fc="w"))
plt.show()
```
